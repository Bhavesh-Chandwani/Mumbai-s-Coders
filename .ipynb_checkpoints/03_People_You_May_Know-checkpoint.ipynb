```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data

def people_you_may_know(user_id, data):
    #First find the friends of the given user.
    user_friends = {}
    for user in data['users']:
        user_friends[user['id']] = set(user['friends'])
        
    if user_id not in user_friends:
        return []
        
    # Taking friends which are directly related to 1
    direct_friends = user_friends[user_id]

    suggestions = {}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends:
                #count mutual friends
                suggestions[mutual] = suggestions.get(mutual, 0) + 1
    # Suggestion.items() converts the dictionary into list of tupes. It says that sort the suggestion using count(value) not the key(id) in descending order.
    sorted_suggestion = sorted(suggestions.items(), key = lambda x:x[1], reverse = True)
    # _ means it does not return the mutual count and it only return the key(id)
    return [user_id for user_id, _ in sorted_suggestion]
        
data = load_data("large_data.json")
# Id of the person we are going to find the mutual friends
user_id = 1
#Recommendation of mutual friends
recommendation = people_you_may_know(user_id, data)
print(f"People you may know for user - {user_id} : {recommendation}")
```
